```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
counts = audi['transmission'].value_counts()
```

#Pie chart and barplot

```python
sns.set_style("whitegrid")
plt.pie(counts,
        labels=counts.index,
        autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
sns.barplot(data = audi,
    x = counts.index,
    y = counts
    )
```
